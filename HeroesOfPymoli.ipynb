```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data["SN"].value_counts())
#total_players

# Display total players in summary table
total_players_table = pd.DataFrame({'Total Players' : [total_players]})
total_players_table
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Unique items
unique_items = len(purchase_data["Item ID"].unique())
#unique_items

# Average price
average_price = purchase_data["Price"].mean()
#average_price

# Number of purchases
number_of_purchases = len(purchase_data["Item Name"])
#number_of_purchases

# Total revenue
total_revenue = purchase_data["Price"].sum()
#total_revenue

# Create summary data frame

purchasing_analysis = pd.DataFrame({"Number of Unique Items" : [unique_items],
                                    "Average Price" : [average_price],
                                    "Number of Purchases" : [number_of_purchases],
                                   "Total Revenue" : [total_revenue]})
#purchasing_analysis

#Display data frame
purchasing_analysis
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchase count of male players
male_purchases = purchase_data.loc[purchase_data["Gender"] == "Male", :]
male_purchase_count = len(male_purchases)
#male_purchase_count

# Average purchase of male players
average_male_purchase = purchase_data.loc[purchase_data["Gender"] == "Male", ["Price"]].mean()
#average_male_purchase

# Purchase total of male players
male_total_purchase = purchase_data.loc[purchase_data["Gender"] == "Male", ["Price"]].sum()
#male_total_purchase

# Purchase count of female players
female_purchases = purchase_data.loc[purchase_data["Gender"] == "Female", :]
female_purchase_count = len(female_purchases)
#female_purchase_count

# Average purchase of female players
average_female_purchase = purchase_data.loc[purchase_data["Gender"] == "Female", ["Price"]].mean()
#average_female_purchase

# Purchase total of female players
female_total_purchase = purchase_data.loc[purchase_data["Gender"] == "Female", ["Price"]].sum()
#female_total_purchase

# Purchase count of other/non-disclosed players
other_purchases = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :]
other_purchase_count = len(other_purchases)
#other_purchase_count

# Average purchase of other/non-disclosed players
average_other_purchase = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", ["Price"]].mean()
#average_other_purchase

# Purchase total of other/non-disclosed players
other_total_purchase = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", ["Price"]].sum()
#other_total_purchase

# Average total purchase per person by gender
average_male_purchase_pp = male_total_purchase / male_players_count
#average_male_purchase_pp
average_female_purchase_pp = female_total_purchase / female_players_count
#average_female_purchase_pp
average_other_purchase_pp = other_total_purchase / other_players_count
#average_other_purchase_pp


# Create summary data frame
purchasing_analysis = pd.DataFrame({
    "": ['Male', 'Female', 'Other / Non-Disclosed'],
    "Purchase Count": [male_purchase_count, female_purchase_count, other_purchase_count],
    "Average Purchase Price": [average_male_purchase, average_female_purchase, average_other_purchase],
    "Total Purchase Value": [male_total_purchase, female_total_purchase, other_total_purchase],
    "Avg Total Purchase per Person": [average_male_purchase_pp, average_female_purchase_pp, average_other_purchase_pp]})

purchasing_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Find min and max ages
# print(purchase_data["Age"].max())
# print(purchase_data["Age"].min())

# Create bins
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]

# Create labels
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorise data and put into bins
pd.cut(purchase_data["Age"], bins=age_bins, labels=age_labels)

# Make a new column Age Ranges
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins=age_bins, labels=age_labels)
#purchase_data

# GroupBy
age_range = purchase_data.groupby("Age Ranges")

# Get number of players in each age range
age_range_totals = age_range["SN"].nunique()

# Calculate percentages of number of players from each age range
age_range_percentage = age_range_totals / total_players * 100


# Create summary data frame
age_demographics = pd.DataFrame({
    "Total Count": age_range_totals, 
    "Percentage of Players": age_range_percentage})

age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{0:,.2f}%".format)

age_demographics.index.name = None

age_demographics
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Groupby
top_spenders = purchase_data.groupby("SN")

# Calculate top spender purchase count
top_spender_purchase_count = top_spenders["Purchase ID"].count()

# Calculate average top spender purchase price
average_top_spender_purchase_price = round(top_spenders["Price"].mean(),2)

# Calculate top spender total purchase value
top_spender_total_purchase_value = top_spenders["Price"].sum()

# Create summary dataFrame
top_spenders_summary = pd.DataFrame({ 
    "Purchase Count": top_spender_purchase_count,
    "Average Purchase Price": average_top_spender_purchase_price,
    "Total Purchase Value": top_spender_total_purchase_value})

top_spenders_summary.sort_values(["Total Purchase Value"], ascending=False).head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Get Item ID name and price columns
items_df = purchase_data[["Item ID", "Item Name", "Price"]]

# GroupBy
popular_items = items_df.groupby(["Item ID","Item Name"])

# Purchase count
item_purchase_count = popular_items["Price"].count()

# Total purchase value
total_item_purchase_value = popular_items["Price"].sum()

# Item price
item_price = total_item_purchase_value / item_purchase_count

# Create summary data frame
most_popular_items = pd.DataFrame({
    "Purchase Count": item_purchase_count, 
    "Item Price": item_price,
    "Total Purchase Value": total_item_purchase_value})

most_popular_items.sort_values('Purchase Count', ascending=False).head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Create summary data frame
most_profitable_items=most_popular_items.sort_values('Total Purchase Value', ascending=False).head()
most_profitable_items
```
